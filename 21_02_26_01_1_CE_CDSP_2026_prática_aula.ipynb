# Variáveis, Tipos e Estrutura de Dados



## **Variáveis**

Em Python, **não temos um comando para definir variáveis**;

As **variáveis** são declaradas no **momento da sua primeira assinatura**;

**Exemplo**:


```python
numero = 10
nome = "João"
print(numero)
print(nome)
```

Em Python, **não precisamos definir previamente os tipos das variáveis**, inclusive a variável **pode mudar de tipo após a sua definição inicial**;

**Exemplo**:

```python
x = 4       # x é do tipo int
print(x)
x = "João" # x agora é do tipo string
print(x)
```

Mas, cadê a saída das variáveis acima?
**print()**

**Tipos de variáveis**

É possível especificar o tipo de uma variável usando uma função adequada:

```python
x = str(3)    # x será '3'
y = int(3)    # y será 3
z = float(3.2)  # z será 3.0
print(x)
print(y)
print(z)
```

# Listas


Listas são uma forma de usarmos múltiplos itens em apenas uma variável. Elas são **ordenáveis** e **mutáveis**.

Exemplo:


```python
minhalista = ["celular", "tablet", "computador"]
print(minhalista)
```

Para saber quantos elementos existem em uma lista nós usamos a função ***len***:


Exemplo:


```python
minhalista = ["celular", "tablet", "computador"]
print(len(minhalista))
```

Para acessar itens em uma lista utilizamos a posição (*index*) no ***array*** que se encontra o elemento

Exemplo:


```python
lista = ["celular", "tablet", "computador"] #Será exibido na tela a palavra "tablet". Pois iniciamos a contagem com zero (em vetores/arrays) em computação.

print(lista[1])
```

E qual a diferença de *array* para lista?

Os ***array*** em python usam menos memória e conseguem processar informações matemáticas mais rapidamente, entretanto os elementos precisam ter **o mesmo tipo de dados**, já nas listas podemos ter diferentes tipos

```python
lista = ["celular", 123, True]

print(lista)
```

Também podemos utilizar um range (intervalo) para acessar os elementos

Exemplo:


```python
lista = ["celular", "tablet", "computador", "relogio", "assistente"]
print(lista[1:4]) #o elemento da 4 posição não está contido
```

E por fim, podemos ocultar um dos elementos do range para que tenhamos todos os elementos iniciando por uma posição, ou todos os elementos até aquela posição
Exemplo:


```python
lista = ["celular", "tablet", "computador", "relogio", "assistente"]
print(lista[:2])
```

```python
lista = ["celular", "tablet", "computador", "relogio", "assistente"]
print(lista[2:])
```

Para **adicionar** itens em uma lista usamos a função append.

Exemplo:


```python
lista = ["celular", "tablet", "computador"]
lista.append("relogio")
print(lista)
```

Para **remover** um elemento usamos a função remove, pelo elemento, ou pop, pelo índice

Exemplo:


```python
lista = ["celular", "tablet", "computador","tablet"]
lista.remove("tablet")
lista.pop(1)
print(lista)
```

## **Condicionais - If**

Para entender como usar condicionais em python precisamos primeiramente entender as condições lógicas matemáticas que podemos utilizar dentro das nossas condições:

* Iguais: a == b
* Não iguais (diferentes): a != b
* Menos que: a < b
* Menor que ou igual a: a <= b
* Maior que: a > b
* Maior que ou igual a: a >= b

Exemplo:

```python
suanota = 10
if suanota > 7:
  print("1. Você está aprovado excelente! Sua nota foi: ", suanota)
#Notem que aqui usamos ',' e não '+' para imprimir o resultado da variável,
#isso acontece porque temos um texto (string) e depois um número inteiro (suanota) para exibir desta forma o '+' daria erro, pois só serve para concatenação/impressão de strings

suanota = 7
if suanota >= 7:
  print("2. Você está aprovado! Sua nota foi: ", suanota)

suanota = 6
if suanota < 7:
  print("3. Você está reprovado! Sua nota foi: ", suanota)

suanota = 10
if suanota == 10:
  print("4. Excelente! Você é um ótimo aluno. Sua nota foi: ", suanota)

```

## **Condicionais - Else**

```python
a = 10
b = 1
if b > a:
  print("b é maior que a")
elif a == b:
  print("a e b são iguais")
else:
  print("a é maior que b")
```

## **Estruturas de repetição - Loop For**

Uma estrutura de repetição bastante utilizada nas linguagens de programação são os **loop** do tipo For.

Um **loop** for é usado para fazer repetições iterativas, seja em listas, tuplas, conjunto de dados, listas, arrays ou em uma string;


```python
equipamentos = ["celular", "computador", "tablet"]
for x in equipamentos:
  print(x)
```

Também é possível ler cada caractere de uma string

Exemplo:


```python
for x in "computador":
  print(x)
```

Podemos **interromper** um laço a qualquer momento com a instrução **break**.

Imagine que estamos pesquisando em uma lista de 1 milhão de pessoas o CPF, e ao encontrarmos queremos parar este loop para que ele não rode todas as pessoas procurando o CPF (temos n formas de fazer isso adequadamente, este é apenas um exemplo).

Exemplo:

```python
minhalista = ["celular", "tablet", "computador"]
print(len(minhalista))
```

```python
pessoas = ["000", "111", "222", "222", "222", "222", "222", "222", "222", "222", "222", "222", "222"]
for x in pessoas:
  print(x)
  if x == "111":
    break
```

Podemos usar a função **range()** para repetir um **For** por **n** interações.
Isto é útil quando, por exemplo você tem que repetir determinado trecho de código código n vezes para uma determinada operação.

Exemplo:


```python
for x in range(10):
  print(x)
```

## **Estruturas de repetição - Loop For aninhado com outros for**

Assim como na instrução If, nós podemos ter ter um for dentro do outro, inclusive com condições If-Else

Exemplo:


```python
config = ["vermelho", "8Gb", "Apple"]
equipamentos = ["computador", "tablet", "celular"]

for x in equipamentos:
  for y in config:
    print(x, y)
```

```python
config = ["vermelho", "8Gb", "Apple"]
equipamentos = ["computador", "tablet", "celular"]

for x in equipamentos:
  for y in config:
    if x=="computador": #filtramos aqui, com uma condição If, apenas a exibição de computador
      print(x, y)
```

## **Estruturas de repetição - Loop While**

```python
i = 0
while i < 10:
  print(i)
  i += 1
```

Também podemos utilizar break dentro de While para parar determinado loop

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

# Funções e métodos

Uma **função** é um bloco de código que só será executado quando for chamado;

Um **método** possui o mesmo princípio. Porém, o método está relacionado a um objeto e por sua vez instanciado de uma classe, é um conceito de orientação a objetos;

Uma **função** pode ter dados passados nela conhecidos como parâmetros e pode retornar um resultado.


```python
def minha_funcao():
  print("Impressão dentro de uma função")
#Note que nada será impresso, pois a função ainda não foi chamada.
```

Para chamar uma função criada basta você chamar esta função usando o nome dela com os parênteses.

Exemplo:


```python
minha_funcao()
```

## **Argumentos**

Podemos passar argumentos para uma função desde que criemos a função para isso, quando a função for chamada teremos que passar um argumento

```python
def minha_funcao(nome):
  print(nome+" Silva")

minha_funcao("João")
minha_funcao("Maria")
minha_funcao("José")
```

Também podemos passar mais de um argumento na função

Exemplo:


```python
def minha_funcao(nome, sobrenome):
  print(nome+" "+sobrenome)

minha_funcao("João", "Pereira")
minha_funcao("Maria", "Silva")
minha_funcao("José", "Silva")
```

Quando não conhecemos a quantidade de argumentos que serão passados usamos *args

Exemplo:


```python
def minha_funcao(*filhos):
  print("O nome dos filhos são: ")
  for filho in filhos:
    print(filho)
minha_funcao("Tomas", "Ana", "Mateus")
```
